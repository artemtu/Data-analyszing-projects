Цель проекта:
получить список лидеров по выручке и количеству стеблей. 

Данные: отчёт по продажам по мастерской. Период с 12.04.2021 по 12.05.2021

```python
df['category'].unique() #получаем уникальные категории товаров
```

```python
df.pivot_table(index='category', values='revenue', aggfunc='sum') #сводная таблица с суммарной выручкой по каждой категории
```

```python
df1 = df.loc[df['category']=='Срезанные цветы' ] #создаем отдельный датафрейм для срезанных цветов
```

```python
display(df1) #проверяем работоспособность нового датафрейма
```

```python
df1.duplicated().sum() #проверяем наличие дубликатов
```

```python
df['product'].value_counts() #ищем одинаковые названия
```

```python
df1['product'].unique() #просматриваем уникальные значения в столбце со срезанными цветами
```

**Предварительный вывод:**

Мы получили огромный список уникальных значений. Это обсусловлено тем, что каждый вид цветка имеет характеристику, такую как высота или цвет. Мы хотим узнать лидера по проданным стеблям и выручке, для этого сорт или высота нам не важны, можем убрать эти данные. 

```python
df1['product_new'] = df1['product'].str.split(' ').str[0] #отбрасываем все значения после первого пробела
```

```python
df1['product_new'].unique() #получаем новый список уникальных значений
```

```python
df1['product_new'].nunique() #получаем количество позиций в новом списке
```

```python
df1['product_new'] = df1['product_new'].str.lower() #делаем все буквы нового списка строчными
```

**Пояснение**

В новом уникальном списке мы увидели несколько видов цветов, которые заведены в базу по-разному. Объединяем их в одну строку.

```python
df1['product_new'] = df1['product_new'].replace('ранункулус', 'ранункулюс') #приводим ранункулусы к ранункулюсам
df1['product_new'] = df1['product_new'].replace('гипсофилла', 'гипсофила') #приводим гипсофиллу к гипсофиле
```

```python
df1['product_new'].unique() #проверяем наши изменения
```

```python
df1.pivot_table(index='product_new', values=['quantity', 'revenue'], aggfunc='sum').sort_values(by='revenue', ascending=False).head(15)#  
#сводная таблица, построенная по виду продукта, количеству проданных стеблей и суммарной выручке. Отфльтрована по убыванию по столбцу "выручка"
#рейтинг цветов по выручке (первые 15 строк)
```

```python
df1.pivot_table(index='product_new', values=['quantity', 'revenue'], aggfunc='sum').sort_values(by='quantity', ascending=False).head(15) #
#сводная таблица, построенная по виду продукта, количеству проданных стеблей и суммарной выручке. Отфльтрована по убыванию по столбцу "количество"
#рейтинг цветов по количеству проданных стеблей (первые 15 строк)
```
